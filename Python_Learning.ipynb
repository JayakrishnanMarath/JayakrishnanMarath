### Legends

### Without Legend

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.7,0.7])
axes.plot(x,y,"r--")
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Bigger_Scale")
axes.plot(y,x,"g--")

plt.show()
```

### With Legend

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.7,0.7])
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Bigger_Scale")
axes.plot(x**2,y,"r--",label="X Squared")
axes.plot(x,y**2,"g--",label="Y Squared")

### We use Legends to identify the lines in the plot

axes.legend(loc=0)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,x**2,"r--",label="X Squared")
axes[0].plot(x,x**3,"g--",label="X Cubed")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].set_title("X Squared")
axes[0].legend(loc=0)
axes[1].plot(x,x**2,"r--",label="X Squared")
axes[1].plot(x,x**3,"g--",label="X Cubed")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")
axes[1].set_title("X Cubed")
axes[1].legend(loc=0)
plt.show()
```
